```python
veri = veri[veri["Weather_Conditions"] != "Unknown"]
```

```python
hava_durumu = veri["Weather_Conditions"].value_counts()
```

```python
hava_durumu
```

```python
hava_durumlari = veri["Weather_Conditions"].unique()
```

```python
hava_durumlari
```

```python
hava_durum_kaza = hava_durumu.values
```

```python
hava_durum_kaza
```

```python
plt.figure(figsize=(20,10) , facecolor = "pink")
sns.barplot(x = hava_durumlari,y = hava_durum_kaza, color = "red")
```

```python
plt.figure(figsize=(20,10) , facecolor = "pink")
plt.pie(hava_durum_kaza, labels= hava_durumlari, colors = sns.color_palette(),startangle = 30,textprops={'size': 'large'},autopct="%1.1f%%")
plt.legend(loc ="lower left")
plt.title("Accident Rate by Weather Conditions",weight="bold")
plt.show()
```
